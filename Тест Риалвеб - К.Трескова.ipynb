```python
pd.options.mode.chained_assignment = None # Отключаем warnings
# Запишем в переменную type_car_new отредактированные типы: Седан - Хетчбек - Кроссовер - Внедорожник

df['type_car_new'] = 0
for i in range(len(df['type_car'])): # Пройдем по всем значениям столбца

    if df['type_car'][i].find('едан') or df['type_car'][i].find('ниверсал') != -1: # Если значение является строкой (не 0, где раньше был Nan) и в
                                                                # нем есть подстрока 'руб.', значит в столбце явно указана сумма гонорара 
        df['type_car_new'][i] = 'Седан'
    if df['type_car'][i].find('тчбек') != -1:
        df['type_car_new'][i] = 'Хетчбек'
    if df['type_car'][i].find('совер') != -1:
        df['type_car_new'][i] = 'Кроссовер'
    if df['type_car'][i].lower().find('внедорожник') != -1:
        df['type_car_new'][i] = 'Внедорожник'
    if df['type_car'][i].lower().find('упе') != -1:
        df['type_car_new'][i] = 'Купе'
    if df['type_car'][i].lower().find('родстер') != -1:
        df['type_car_new'][i] = 'Родстер'
    if df['type_car'][i].lower().find('купе') != -1:
        df['type_car_new'][i] = 'Купе'
    if df['type_car'][i].lower().find('вэн') != -1:
        df['type_car_new'][i] = '(Мини)вэн'
    if df['type_car'][i].lower().find('пикап') != -1:
        df['type_car_new'][i] = 'Пикап'
    if df['type_car'][i].lower().find('кабрио') != -1:
        df['type_car_new'][i] = 'Кабриолет'                
```

```python
filter_car = (df['type_car_new'] == 'Кроссовер') | (df['type_car_new'] == 'Седан')| (df['type_car_new'] == 'Внедорожник')| (df['type_car_new'] == 'Хетчбек')
```

```python
df[filter_car] # Отобразим таблицу только по 4 основным кузовам: Седан - Хетчбек - Кроссовер - Внедорожник
```

```python
# Типу кузова присвоим категорию для дальнейшей кодировки всего датафрейма, что позволит быстрее обработать информацию
def getCarCategory(arg):

    if(arg == 'Седан'):
        out = 1
    elif(arg == 'Хетчбек'):
        out = 2
    elif(arg == 'Кроссовер'):
        out = 3
    elif(arg == 'Внедорожник'):
        out = 4
    elif(arg in ['(Мини)вэн', 'Купе', 'Пикап', 'Родстер', 'Кабриолет']):
        out = 0
    
    return list(utils.to_categorical(out, 5).astype('int'))

    
```
