```python
# 1.      Рассчитать индексы сезонности по месяцам 2014-2015 в разрезе групп городов

import numpy as np
import pandas as pd
df = pd.read_excel('../Public/Documents/file.xlsx', sheet_name='История')
# df.isnull().sum()

# Делим таблицу на две по годам: 
df14 = df.filter(regex='2014', axis=1)
df15 = df.filter(regex='2015', axis=1)

dfsort = df[['ГруппаГородов""','Наименование']]
df14 = pd.concat([dfsort, df14], axis=1)
df15 = pd.concat([dfsort, df15], axis=1)

# df['ГруппаГородов""'].value_counts()
df['ГруппаГородов""'] = df['ГруппаГородов""'].replace({'группа юг': 'Группа Юг'})

# df.dropna(inplace=True) # вопрос!....если нужно чистить базу от пустых значений для первого задания, то можно пересчитать с этой строки
groups_mean_month = df.groupby('ГруппаГородов""').mean()
# groups_mean_month
groups_mean_month14 = groups_mean_month.iloc[:,:12]
groups_mean_month14.set_axis(list(range(1,13)), axis=1, inplace=True)
# groups_mean_month14.head(2)
groups_mean_month15 = groups_mean_month.iloc[:,12:]
groups_mean_month15.set_axis(list(range(1,13)), axis=1, inplace=True)
# groups_mean_month15.head(2)

df_groups_mean_month = pd.concat([groups_mean_month14, groups_mean_month15], axis=0)
# df_groups_mean_month[df_groups_mean_month.index=='Группа Ессентуки']
df_groups_mean_month_result = df_groups_mean_month.groupby('ГруппаГородов""').mean()
df_groups_mean_month_result
```

```python
# 2.      Вычислить прирост по сопоставимым объектам 2015 к 2014 итого и в разрезе групп городов
# Сопоставимыми объектами считаются объекты, которые работали в рассматриваемый и сравниваемый периоды

df = pd.read_excel('../Public/Documents/file.xlsx', sheet_name='История')
df['ГруппаГородов""'] = df['ГруппаГородов""'].replace({'группа юг': 'Группа Юг'})
dfsort = df[['ГруппаГородов""','Наименование']]
df14 = df.filter(regex='2014', axis=1)
df14 = pd.concat([dfsort, df14], axis=1)
df14.set_index(['ГруппаГородов""','Наименование'], drop=True, inplace=True)
df14.dropna(axis=0, how='all', inplace=True)
df15 = df.filter(regex='2015', axis=1)
df15 = pd.concat([dfsort, df15], axis=1)
new_df = df15.join(df14, on=['ГруппаГородов""','Наименование'], how="inner")


new_df14 = new_df.filter(regex='2014', axis=1)
new_df15 = new_df.filter(regex='2015', axis=1)
dfsort = new_df[['ГруппаГородов""','Наименование']]
new_df14 = pd.concat([dfsort, new_df14], axis=1)
new_df15 = pd.concat([dfsort, new_df15], axis=1)

new_df14['sum14'] = new_df14.sum(axis=1) 
new_df15['sum15'] = new_df15.sum(axis=1) 

new_df = pd.concat([new_df14, new_df15], axis=1)
new_df['growth %'] = round((new_df['sum15']-new_df['sum14'])/new_df['sum14']*100,2)
new_df_growth = new_df[['ГруппаГородов""','Наименование','growth %']]

# sum(new_df_growth.iloc[:,2]==new_df_growth.iloc[:,3]) 
# sum(new_df_growth.iloc[:,0]==new_df_growth.iloc[:,1]) 
cols = [1,3,4]
new_df_growth = new_df_growth.iloc[:, cols]
```
